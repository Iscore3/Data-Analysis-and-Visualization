### Read data

```python
import pandas as pd
link = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
df = pd.read_csv(link, header=None)
```

```python
df.head()
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns=headers
df.head()
```

### Drop missing values

```python
df.dropna(subset=["price"], axis=0).head()
```

### Save dataframe to excel (csv)

```python
df.to_csv("example.csv", index=False)
```

### Basic analysis of Dataframe

```python
df.dtypes.head()
```

```python
df.describe()
```

```python
df.info # result is long
```

### Replace values

```python
import numpy as np
df.replace("?", np.nan, inplace = True)
df.head(5)
```

### Search for missing data

```python
missing_data = df.isnull()
missing_data.head(5)
```

### Calculate the mean of a column

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
avg_norm_loss
```

### Count the values in a columns

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
drive_wheels_counts = df['drive-wheels'].value_counts().to_frame()
drive_wheels_counts.rename(columns={'drive-wheels': 'value_counts'}, inplace=True)
drive_wheels_counts.index.name = 'drive-wheels'
drive_wheels_counts
```

### Reset Index

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

### Convert the data type of a column

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
```

### Binning

```python
bins = np.linspace(min(df["city-mpg"]), max(df["city-mpg"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
df['city-mpg-binned'] = pd.cut(df['city-mpg'], bins, labels=group_names, include_lowest=True ) # in which bin belongs to
```

### Dummy variable

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
df = pd.concat([df, dummy_variable_1], axis=1) # icluded 3 times because the code was executed 3 times
#df.drop("fuel-type", axis = 1, inplace=True)
df.head()
```

### Grouping data

```python
df['drive-wheels'].unique()
```

```python
df_group_one = df[['drive-wheels','city-mpg']]
```

```python
df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean()
df_group_one
```

### Correlation

```python
df.corr()
```

```python
from scipy import stats
pearson_coef, p_value = stats.pearsonr(df['city-mpg'], df['highway-mpg'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value) 
```
